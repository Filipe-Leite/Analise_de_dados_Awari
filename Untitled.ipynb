# PRODUZIDOS NO BRASIL ORDENADOS POR AVALIAÃÃO

```python
brazil_only_order_by_imdb = pandasql.sqldf('''SELECT Title, IMDb, Country 
                                                  FROM df 
                                                      WHERE Country = 'Brasil' 
                                                         or Country = 'brasil'
                                                         or Country = 'BRASIL'
                                                         or Country = 'Brazil'
                                                         or Country = 'brazil'
                                                         or Country = 'BRAZIL'
                                                                ORDER by IMDb DESC''')
```

# DIRETORES COM MAIOR MÃDIA EM TÃTULOS DISPONÃVEIS NA NETFLIX E PRIMEVIDEO AO MESMO TEMPO

```python
imdbOrderByDirectors = pandasql.sqldf('''SELECT directors,
                                                imdb,
                                                sum(imdb)/count(directors) AS mean
                                                FROM df
                                                    WHERE netflix = '1'
                                                        AND prime_video = '1' 
                                                            GROUP BY directors
                                                                ORDER BY mean DESC''')
```
